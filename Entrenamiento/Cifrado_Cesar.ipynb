```python
import pandas as pd

alfabeto = "abcdefghijklmnñopqrstuvwxyz"

def encriptar_cesar_es(texto, desplazamiento):
  resultado = ""
  for caracter in texto:
    if caracter.lower() in alfabeto:
      es_mayus = caracter.isupper()
      indice = alfabeto.index(caracter.lower())
      nuevo_indice = (indice + desplazamiento) % len(alfabeto)
      nuevo_caracter = alfabeto[nuevo_indice]
      resultado += nuevo_caracter.upper() if es_mayus else nuevo_caracter
    else:
      resultado += caracter
  return resultado

def desencriptar_cesar_es(texto_cifrado, desplazamiento):
  return encriptar_cesar_es(texto_cifrado, -desplazamiento)

#listas de mensajes y claves
mensaje = ["hola","sol","caracter","paso","llave","corregir","Matenimiento1","loco","cable","seguridad"];
clave = [2,4,5,20,11,8,9,1,13,6];
cifrado = [];
desifrado = [];

for i in range(len(mensaje)):
  cifrado.append(encriptar_cesar_es(mensaje[i],clave[i]))
print(cifrado)

for i in range(len(cifrado)):
  desifrado.append(desencriptar_cesar_es(cifrado[i],clave[i]))
print(desifrado)

data_set = {"msj_original": mensaje,
            "clave": clave,
            "msj_cifrado": cifrado,
            "msj_desifrado": desifrado}

print(data_set)

df = pd.DataFrame(data_set)
print(df)
df
```
