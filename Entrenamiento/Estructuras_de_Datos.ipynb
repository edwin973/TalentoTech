```python
paises = ["Colombia", "Francia", "Estados Unidos", "Egipto", "Japon", "Nigeria", "Rusia", "Argentina", "Holanda", "Espa単a"] #Definicion de una lista

animales = ("\nGato","Raton", "Vaca", "Leon", "Jirafa", "Serpiente", "Leopardo", "Aguila", "Ping端ino", "Armadillo") #Definicion de una tupla

objetos = {"\nCuchillo", "Estante", "Martillo", "Lapiz", "Pantalla", "Puerta", "Cama", "Libro", "Zapato", "Cepillo"} #Definicion de un conjunto

palabras = {  #Definicion de un diccionario
  "\nbook":"libro",
  "pain":"dolor",
  "yelllow":"amarillo",
  "pencil":"lapiz",
  "phone":"celular",
  "door":"puerta",
  "car":"carro",
  "cat":"gato",
  "desk":"escritorio",
  "police":"policia"
}

#imprimir los elementos de la lista
for pais in paises:
    print(pais)

#imprimir los elementos de la tupla
for animal in animales:
    print(animal)

#imprimir los elementos del conjunto
for objeto in objetos:
    print(objeto)

#imprimir los elementos del diccionario
for llave, valor in palabras.items():
    print(llave, " -> " , valor)
```
