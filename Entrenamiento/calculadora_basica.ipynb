```python
print("Calculadora Basica")
print("")
print("1. Suma")
print("2. Resta")
print("3. Multiplicacion")
print("4. Division")
print("5. Potencia")

opcion=int(input("Ingrese la operación que desea realizar(Opcion numerica):"))
print("")
prinumero=float(input("Ingrese el primer número:"))
segnumero=float(input("Ingrese el segundo número:"))

if opcion>= 1 and opcion <= 5:
  if opcion==1:
    resultado = prinumero + segnumero
    print("El resultado de la suma es: ", resultado)
  if opcion==2:
    resultado = prinumero - segnumero
    print("El resultado de la resta es: ", resultado)
  if opcion==3:
    resultado = prinumero * segnumero
    print("El resultado de la multiplicación es: ", resultado)
  if opcion==4:
    if segnumero == 0:
      print("No se puede realizar la operación con los datos proporcionados")
    else:
      resultado = prinumero / segnumero
      print("El resultado de la division es: ", resultado)
  if opcion==5:
    resultado = prinumero ** segnumero
    print("El resultado de la potencia es: ", resultado)
else:
  print("Se ha ingresado el valor para una operación inexistente, vuelve a intentarlo")
```
