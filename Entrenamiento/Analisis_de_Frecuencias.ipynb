```python
from collections import Counter

def analizar_frecuencia_caracteres(texto):

    # Convertir a minúsculas y eliminar espacios para un conteo uniforme
    texto_limpio = texto.lower().replace(" ", "")

    # Usar Counter para contar las ocurrencias de cada caracter
    frecuencias = Counter(texto_limpio)

    # Calcular el total de caracteres
    total_caracteres = sum(frecuencias.values())

    print(f"--- Análisis de Frecuencia de Caracteres ---")
    print(f"Texto original: '{texto}'")
    print(f"Total de caracteres limpios: {total_caracteres}\n")

    print("Frecuencia Absoluta:")
    for caracter, cuenta in sorted(frecuencias.items()):
        print(f"'{caracter}': {cuenta} veces")

    print("\nFrecuencia Relativa (Proporción):")
    for caracter, cuenta in sorted(frecuencias.items()):
        proporcion = (cuenta / total_caracteres) * 100 if total_caracteres > 0 else 0
        print(f"'{caracter}': {proporcion:.2f}%")


mi_texto = "3# 16_@!5?6#=_# 2> 23 #6!2 |2 2>16_$_6 15% 13#72 >2162!# 5 |2 /% 45|5 2%_?4#!_15"
analizar_frecuencia_caracteres(mi_texto)
```

```python
import random

def ahorcado(palabra_secreta):

    palabra_guiones = ["_"] * len(palabra_secreta)

    intentos_maximos = 5
    intentos = 0
    letras_adivinadas = set()

    print("Bienvenido al juego del Ahorcado!")
    print(f"La palabra tiene {len(palabra_secreta)} letras.")
    print("Tienes que adivinar la palabra antes de que se acaben los intentos.")

    # Comienza el juego
    while intentos < intentos_maximos:
        print("\nPalabra: " + " ".join(palabra_guiones))
        print(f"Tienes {intentos_maximos - intentos} intentos restantes.")
        print(f"Letras adivinadas: {', '.join(letras_adivinadas)}")

        # Entrada del jugador
        letra = input("Adivina una letra: ").lower()

        # Validación de la letra
        if len(letra) != 1 or not letra.isalpha():
            print("Por favor, ingresa solo una letra.")
            continue

        if letra in letras_adivinadas:
            print(f"Ya adivinaste la letra '{letra}'. Intenta con otra.")
            continue

        letras_adivinadas.add(letra)

        if letra in palabra_secreta:
            print(f"¡Correcto! La letra '{letra}' está en la palabra.")
            # Actualiza los guiones con la letra adivinada
            for i in range(len(palabra_secreta)):
                if palabra_secreta[i] == letra:
                    palabra_guiones[i] = letra
        else:
            print(f"¡Incorrecto! La letra '{letra}' no está en la palabra.")
            intentos += 1

        # Verifica si el jugador ha adivinado la palabra completa
        if "_" not in palabra_guiones:
            print(f"\n¡Felicidades! Has adivinado la palabra: {''.join(palabra_guiones)}")
            break

    if intentos == intentos_maximos:
        print(f"\n¡Te has quedado sin intentos! La palabra era: {palabra_secreta}")

palabra_secreta = input("Ingresa la palabra secreta: ")
ahorcado(palabra_secreta)

if __name__ == "__main__":
    ahorcado(palabra_secreta)


```
