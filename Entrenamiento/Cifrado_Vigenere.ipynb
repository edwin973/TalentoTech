```python
# Función para cifrar un texto usando el cifrado de Vigenère
def cifrado_vigenere(texto, clave):
  texto_cifrado = ""
  clave_repetida = (clave * (len(texto) // len(clave))) + clave[:len(texto) % len(clave)]

  for i in range(len(texto)):
    letra_texto = texto[i].upper()
    letra_clave = clave_repetida[i].upper()

    if 'A' <= letra_texto <= 'Z':
      # Calcular el desplazamiento
      desplazamiento = ord(letra_clave) - ord('A')

      # Aplicar el desplazamiento
      letra_cifrada_ascii = ord('A') + (ord(letra_texto) - ord('A') + desplazamiento) % 26
      letra_cifrada = chr(letra_cifrada_ascii)
      texto_cifrado += letra_cifrada
    else:
      # Si no es una letra, agregarla tal cual
      texto_cifrado += texto[i]

  return texto_cifrado

# Función para descifrar un texto usando el cifrado de Vigenère
def descifrado_vigenere(texto_cifrado, clave):
  texto_original = ""
  clave_repetida = (clave * (len(texto_cifrado) // len(clave))) + clave[:len(texto_cifrado) % len(clave)]

  for i in range(len(texto_cifrado)):
    letra_cifrada = texto_cifrado[i].upper()
    letra_clave = clave_repetida[i].upper()

    if 'A' <= letra_cifrada <= 'Z':
      # Calcular el desplazamiento
      desplazamiento = ord(letra_clave) - ord('A')

      # Aplicar el desplazamiento inverso
      letra_original_ascii = ord('A') + (ord(letra_cifrada) - ord('A') - desplazamiento + 26) % 26
      letra_original = chr(letra_original_ascii)
      texto_original += letra_original
    else:
      # Si no es una letra, agregarla tal cual
      texto_original += texto_cifrado[i]

  return texto_original

# Ejemplo de uso
texto_a_cifrar = input("Ingrese el mensaje a cifrar")
clave_secreta = input("Ingrese la clave")

texto_cifrado_resultante = cifrado_vigenere(texto_a_cifrar, clave_secreta)
print(f"Texto a cifrar: {texto_a_cifrar}")
print(f"Clave: {clave_secreta}")
print(f"Texto cifrado: {texto_cifrado_resultante}")

texto_descifrado_resultante = descifrado_vigenere(texto_cifrado_resultante, clave_secreta)
print(f"Texto cifrado: {texto_cifrado_resultante}")
print(f"Clave: {clave_secreta}")
print(f"Texto descifrado: {texto_descifrado_resultante}")
```
