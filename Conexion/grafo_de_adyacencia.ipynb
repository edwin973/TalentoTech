```python
# CREACION DE UN GRAFO DE ADYACENCIA DE MANERA MANUAL
import networkx as nx
import matplotlib.pyplot as plt

# Crear un grafo vacío
G = nx.Graph()

# Agregar 4 nodos
G.add_nodes_from(["v1", "v2", "v3", "v4"])

# Agregar algunas aristas (conexiones) entre los nodos
G.add_edge("v1", "v2", label='a1')
G.add_edge("v2", "v1", label='a2')
G.add_edge("v2", "v3", label='a3')
G.add_edge("v3", "v4", label='a4')
G.add_edge("v4", "v1", label='a5')

# Posiciones de los nodos (para un layout visual estable)
pos = nx.spring_layout(G)

# Dibujar el grafo
nx.draw(G, pos, with_labels=True, node_color='lightblue', node_size=1000, font_weight='bold')

# Extraer etiquetas de las aristas
edge_labels = nx.get_edge_attributes(G, 'label')

# Dibujar etiquetas en las aristas
nx.draw_networkx_edge_labels(G, pos, edge_labels=edge_labels, font_color='red', label_pos=0.5)
plt.title("Grafo con 4 nodos")
plt.show()
```

```python
import networkx as nx
import matplotlib.pyplot as plt

# Crear un grafo dirigido
G = nx.DiGraph()

# Agregar 4 nodos
G.add_nodes_from(["v1", "v2", "v3", "v4"])

# Agregar aristas dirigidas entre los nodos
G.add_edge("v1", "v2", label='a1')
G.add_edge("v2", "v1", label='a2')
G.add_edge("v2", "v3", label='a3')
G.add_edge("v3", "v4", label='a4')
G.add_edge("v4", "v1", label='a5')

# Posiciones de los nodos
pos = nx.spring_layout(G)

# Dibujar el grafo
nx.draw(G, pos, with_labels=True, node_color='lightblue', node_size=1000, font_weight='bold', arrows=True)

# Extraer y dibujar etiquetas de las aristas
edge_labels = nx.get_edge_attributes(G, 'label')
nx.draw_networkx_edge_labels(G, pos, edge_labels=edge_labels, font_color='red', label_pos=0.3)

plt.title("Grafo dirigido con aristas a1 y a2 diferenciadas")
plt.show()
```

```python
# CREACION DE UN GRAFO DE ADYACENCIA Y LA MATRIZ CORRESPONDIENTE DE MANERA AUTOMATICA

import networkx as nx
import numpy as np
import matplotlib.pyplot as plt

def crear_grafo_y_matriz(x):
    # Crear un grafo vacío (puedes usar nx.DiGraph() si quieres que sea dirigido)
    G = nx.Graph()

    # Crear nodos automáticamente: v1, v2, ..., vx
    nodos = [f'v{i+1}' for i in range(x)]
    G.add_nodes_from(nodos)

    # Aquí conectamos cada nodo con el siguiente para formar un ciclo
    for i in range(x):
        G.add_edge(nodos[i], nodos[(i+1) % x])

    # Dibujar el grafo
    pos = nx.circular_layout(G)
    nx.draw(G, pos, with_labels=True, node_color='lightblue', node_size=800, font_weight='bold')
    plt.title(f"Grafo con {x} nodos")
    plt.show()

    # Obtener matriz de adyacencia como array de NumPy
    matriz_adyacencia = nx.to_numpy_array(G, nodelist=nodos, dtype=int)

    # Mostrar matriz de adyacencia
    print("Matriz de adyacencia:")
    print(matriz_adyacencia)

    return matriz_adyacencia


crear_grafo_y_matriz(4)
```
