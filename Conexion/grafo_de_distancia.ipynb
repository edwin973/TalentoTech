```python
import networkx as nx
import matplotlib.pyplot as plt

# Matriz de distancias
dist_matrix = [
    [0, 1, 2, 3],
    [1, 0, 1, 2],
    [2, 3, 0, 1],
    [1, 2, 3, 0]
]

def crear_grafo_desde_matriz(matriz):
    G = nx.Graph()
    n = len(matriz)

    for i in range(n):
        for j in range(n):
            if matriz[i][j] != 0:
                G.add_edge(i, j, weight=matriz[i][j])

    return G

def mostrar_matriz_adyacencia(G):
    A = nx.to_numpy_array(G, weight='weight')
    print("Matriz de adyacencia basada en pesos:")
    print(A)

def dibujar_grafo(G):
    pos = nx.spring_layout(G)
    edge_labels = nx.get_edge_attributes(G, 'weight')
    nx.draw(G, pos, with_labels=True, node_color='skyblue', node_size=700, font_weight='bold')
    nx.draw_networkx_edge_labels(G, pos, edge_labels=edge_labels)
    plt.title("Grafo generado desde matriz de distancias")
    plt.show()

# Crear grafo
G = crear_grafo_desde_matriz(dist_matrix)

# Mostrar matriz
mostrar_matriz_adyacencia(G)

# Dibujar grafo
dibujar_grafo(G)
```
